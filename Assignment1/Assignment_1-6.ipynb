```python
import pandas as pd
def get_xG(match_ids, exclude_team_name):
    """
    Extracts and processes shot data for specified matches, excluding a specific team, to calculate expected goals (xG).

    Parameters:
    - match_ids (list of int): A list of match IDs.
    - exclude_team_name (str): The name of the team to exclude (default is "Spain").

    Returns:
    - pd.DataFrame: A DataFrame containing shot data with columns for x, y, xG, outcome, and match_id.
    """
    all_shots = pd.DataFrame()  # Initialize an empty DataFrame to collect data

    # Iterate over each match ID
    for match_id in match_ids:
        # Load events for the specified match
        events = sb.events(match_id=match_id)
        
        # Filter for shot events excluding the specified team
        shots = events[(events['type'] == 'Shot') & (events['team'] != exclude_team_name)].copy()
        
        # Select and rename necessary columns
        shots = shots[['location', 'shot_statsbomb_xg', 'shot_outcome']]
        shots['x'] = shots['location'].apply(lambda loc: loc[0])
        shots['y'] = shots['location'].apply(lambda loc: loc[1])
        shots['xG'] = shots['shot_statsbomb_xg']
        shots['outcome'] = shots['shot_outcome'].apply(lambda x: 'saved' if x != 'Goal' else 'goal')
        shots['match_id'] = match_id  # Store match_id in the DataFrame

        # Append the shots data for this match to the all_shots DataFrame
        all_shots = pd.concat([all_shots, shots], ignore_index=True)
    
    # Drop columns no longer needed
    all_shots.drop(columns=['location', 'shot_statsbomb_xg', 'shot_outcome'], inplace=True)

    return all_shots
```

```python
match_ids_spain = [3906390,3904628,3893806,3893822,3902240,3901733,3893791]
xG_spain= get_xG(match_ids=match_ids_spain,exclude_team_name="Spain Women's")
print(xG_spain["xG"].sum()/7)
xG_spain
```
