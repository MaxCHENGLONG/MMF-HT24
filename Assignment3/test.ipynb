```python
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.cm import get_cmap
from mplsoccer import Pitch

"""
    Investigate the types of runs made by players during possession chains that entered the final third.
"""

# 设置比赛的 ID，CL 决赛
match_id = 18768058

# 加载 wyscout 事件数据 - 坐标已转换为 opta 格式
df_events = pd.read_parquet(f"Data/{match_id}_stories.parquet")

# 加载预生成的 "runs" 数据
df_runs = pd.read_parquet(f"Data/{match_id}_runs.parquet")

# 添加一列表示跑动是否在控球链中
df_runs['In possession runs'] = df_runs['possession_id'].isin(df_events['possession_id'].unique())

# 筛选出进入前场三分之一的所有控球链
final_third_possessions = df_events[df_events['end_x'] > 66]  # 前场三分之一在 x > 66 的区域
possession_ids_final_third = final_third_possessions['possession_id'].unique()

# 筛选出在这些控球链中的所有跑动
df_final_third_runs = df_runs[df_runs['possession_id'].isin(possession_ids_final_third)]

# 绘制所有进入前场三分之一的跑动
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))
pitch = Pitch(pitch_type="opta",  # 使用 opta 类型的球场
              goal_type='box',  # 使用矩形球门
              pitch_color="w",  # 设置球场颜色为白色
              linewidth=1,  # 设置线条宽度为 1
              spot_scale=0,  # 不绘制点球点
              line_color="k",  # 线条颜色为黑色
              line_zorder=1)  # 设置线条的绘制顺序

# 绘制球场
pitch.draw(ax)

# 创建颜色集
name = "Set1"  # 选择颜色集名称
cmap = get_cmap(name)  # 获取颜色映射
colors = cmap.colors  # 获取颜色列表

# 初始化变量以保存跑动类型数据
forward_runs = 0
lateral_runs = 0
backward_runs = 0

# 绘制所有跑动
for idx, row in df_final_third_runs.iterrows():
    # 计算跑动方向的角度
    dx = row['end_x'] - row['start_x']
    dy = row['end_y'] - row['start_y']
    angle = np.arctan2(dy, dx)  # 使用 arctan2 计算方向角
    angle_degrees = np.degrees(angle)

    # 确定跑动的类型
    if -45 <= angle_degrees <= 45:
        run_type = 'forward'
        forward_runs += 1
    elif 45 < angle_degrees < 135 or -135 < angle_degrees < -45:
        run_type = 'lateral'
        lateral_runs += 1
    else:
        run_type = 'backward'
        backward_runs += 1

    # 绘制跑动箭头
    pitch.arrows(row['start_x'], row['start_y'],  # 起点坐标
                 row['end_x'], row['end_y'],  # 终点坐标
                 width=1.8,  # 箭头宽度
                 headwidth=3, headlength=3, headaxislength=2,  # 箭头的头部尺寸
                 color=colors[idx % len(colors)],  # 使用颜色列表循环颜色
                 edgecolor="k",  # 设置边框颜色为黑色
                 linewidth=1,  # 边框宽度
                 alpha=0.8,  # 透明度
                 zorder=3,  # 设置绘制顺序
                 label=row['player'] if idx == 0 else "",  # 只为第一个跑动添加球员标签
                 ax=ax)

# 添加标题
ax.set_title("Runs made by players during possession chains entering the final third")

# 显示不同类型跑动的统计数据
plt.figtext(0.15, 0.01, f"Forward Runs: {forward_runs}\nLateral Runs: {lateral_runs}\nBackward Runs: {backward_runs}", fontsize=10)

# 保存输出图像
os.makedirs("outputs", exist_ok=True)  # 如果不存在，则创建 "outputs" 目录
#fig.savefig(f"outputs/final_third_runs_types.jpg", format='jpg', dpi=200, facecolor=fig.get_facecolor())

plt.show()  # 显示图像
```

```python
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
from mplsoccer import Pitch

"""
    Plot all runs of a selected player to analyze their run patterns.
    Characterize the average run direction and the variation in the player's run direction with a single metric.
"""

# 设置比赛的 ID，CL 决赛
match_id = 18768058
# 设置需要分析的球员名称
selected_player_id = 'home_team_player_9'  # 替换为球员的真实名字 # Erling Haaland

# 加载 wyscout 事件数据 - 坐标已转换为 opta 格式
df_events = pd.read_parquet(f"Data/{match_id}_stories.parquet")

# 加载预生成的 "runs" 数据
df_runs = pd.read_parquet(f"Data/{match_id}_runs.parquet")

# 筛选出所选球员的所有跑动数据
df_player_runs = df_runs[(df_runs['player_id'] == selected_player_id) ] #& (df_runs['In possession runs'] == True)

# 打印筛选出的球员跑动数据
print(df_player_runs)

# 绘制所有跑动
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))
pitch = Pitch(pitch_type="opta",  # 使用 opta 类型的球场
              goal_type='box',  # 使用矩形球门
              pitch_color="w",  # 设置球场颜色为白色
              linewidth=1,  # 设置线条宽度为 1
              spot_scale=0,  # 不绘制点球点
              line_color="k",  # 线条颜色为黑色
              line_zorder=1)  # 设置线条的绘制顺序

# 绘制球场
pitch.draw(ax)

# 初始化变量以保存跑动类型数据
forward_runs = 0
lateral_runs = 0
backward_runs = 0
angles = []

# 绘制球员的跑动并区分不同类型
for idx, row in df_player_runs.iterrows():
    dx = row['end_x'] - row['start_x']
    dy = row['end_y'] - row['start_y']
    angle = np.arctan2(dy, dx)
    angle_degrees = np.degrees(angle)
    angles.append(angle_degrees)
    
    # 确定跑动的类型并设置颜色
    if -45 <= angle_degrees <= 45:
        run_type = 'forward'
        color = 'g'  # 前向跑动使用绿色
        forward_runs += 1
    elif 45 < angle_degrees < 135 or -135 < angle_degrees < -45:
        run_type = 'lateral'
        color = 'y'  # 横向跑动使用黄色
        lateral_runs += 1
    else:
        run_type = 'backward'
        color = 'r'  # 后向跑动使用红色
        backward_runs += 1

    # 绘制跑动箭头
    pitch.arrows(row['start_x'], row['start_y'],  # 起点坐标
                 row['end_x'], row['end_y'],  # 终点坐标
                 width=1.8,  # 箭头宽度
                 headwidth=3, headlength=3, headaxislength=2,  # 箭头的头部尺寸
                 color=color,  # 根据跑动类型设置颜色
                 edgecolor="k",  # 设置边框颜色为黑色
                 linewidth=1,  # 边框宽度
                 alpha=0.8,  # 透明度
                 zorder=3,  # 设置绘制顺序
                 ax=ax)

# 设置图标题
ax.set_title(f"Erling Haaland Runs (All runs)")
plt.show()  # 显示图像

# 打印不同类型跑动的统计数据
print(f"Forward Runs: {forward_runs}")
print(f"Lateral Runs: {lateral_runs}")
print(f"Backward Runs: {backward_runs}")

# 可视化跑动类型的统计
fig, ax = plt.subplots(figsize=(10, 7))
categories = ['Forward Runs', 'Lateral Runs', 'Backward Runs']
values = [forward_runs, lateral_runs, backward_runs]
ax.bar(categories, values, color=['g', 'y', 'r'])
ax.set_title(f"Erling Haaland Run Types Counts (All runs)")
ax.set_ylabel("Number of Runs")
ax.set_yticks(range(0, max(values) + 2, 1))
plt.show()

# 计算平均跑动方向和方向的方差
avg_direction = np.mean(angles)
variation = np.std(angles)

# 打印平均跑动方向和方向的方差
print(f"Average Run Direction: {avg_direction:.2f} degrees")
print(f"Variation in Run Direction: {variation:.2f} degrees")
```
